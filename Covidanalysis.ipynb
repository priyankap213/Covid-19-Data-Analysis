```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
print("Mdoules are imported")
                                       #ALl Modules are imported
```

```python
corona_csv = pd.read_csv("corona.csv")                  #dataset is read using read_csv from pandas
corona_csv.head(10)                                     #head is use to show contents of dataset
```

```python
corona_csv.drop(["Lat","Long"],axis=1,inplace=True)       #drop method accepts a list of names to drop , axis =1 means columns , inplace means no different copy will be generated
corona_agg = corona_csv.groupby("Country/Region").sum()    #to group data by sinle country and not by different regions
corona_agg.head(10)
```

```python
corona_agg.loc["China"].plot()     #loc access no of columns for that row i.e china, plot gives a graph
corona_agg.loc["India"].plot()
plt.legend()                       # gives name to the lines
```

```python
corona_agg.loc["China"][:4].plot()
```

```python
corona_agg.loc["India"].diff().plot()     #diff() calcultes difference btwn series of nos so there it shows the rate at which it increases
```

```python
corona_agg.loc["India"].diff().max()          #max cases in a day #21june2020
```

```python
countries = list(corona_agg.index)
max_infection =[]
for c in countries:
    max_infection.append(corona_agg.loc[c].diff().max())
max_infection
```

```python
corona_agg["max_infection_rates"] = max_infection
corona_agg.head()
```

```python
corona_data = pd.DataFrame(corona_agg["max_infection_rates"]) #dtaframe is used to create new dataset ,with column max.. and index as countries
corona_data.head()
```

```python
happiness_csv = pd.read_csv("happiness.csv")
happiness_csv.head()
```

```python
happiness_csv.drop(["Overall rank","Score","Generosity","Perceptions of corruption"],axis=1,inplace=True) #dont exe two times
happiness_csv.head()
```

```python
happiness_csv.set_index("Country or region",inplace=True)  #set_index to set a column as index
#happiness_csv.head()
```
